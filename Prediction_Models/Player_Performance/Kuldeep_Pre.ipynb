```python
import pandas as pd
from sklearn.model_selection import KFold, cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
```

```python
# Load the data from the CSV file
data = pd.read_csv("Kuldeep.csv")
```

```python
# Separate features (X) from the target variable (y)
X = data.drop('Wickets', axis=1)
y = data['Wickets']

labelencoder_1 = LabelEncoder()
X['Opposition'] = labelencoder_1.fit_transform(X['Opposition'])

labelencoder_2 = LabelEncoder()
X['Ground'] = labelencoder_2.fit_transform(X['Ground'])

labelencoder_3 = LabelEncoder()
X['Innings'] = labelencoder_3.fit_transform(X['Innings'])

labelencoder_4 = LabelEncoder()
X['Overs'] = labelencoder_4.fit_transform(X['Overs'])

#labelencoder_5 = LabelEncoder()
#X['Economy'] = labelencoder_5.fit_transform(X['Economy'])

# labelencoder_6 = LabelEncoder()
# X['Maidens'] = labelencoder_6.fit_transform(X['Maidens'])

X = X[['Opposition','Ground','Innings','Overs']]

# K-fold cross-validation
kf = KFold(n_splits=10, shuffle=True, random_state=42)
mse_scores = []

for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Train the model on the training data
    model = RandomForestRegressor(random_state=42)
    model.fit(X_train, y_train)

    # Make predictions on the test data
    y_pred = model.predict(X_test)

    # Calculate the mean squared error
    mse = mean_squared_error(y_test, y_pred)
    mse_scores.append(mse)

# Calculate the average mean squared error
avg_mse = sum(mse_scores) / len(mse_scores)
print("Average Mean Squared Error:", avg_mse)

Opposition = input("Enter the Opposition : ")
Ground = input("Enter the Ground : ")
Innings = input("Enter the Innings : ")
Overs = input("Enter the Avg_Overs : ")
#Economy = input("Enter Avg_Economy : ")

new_data = {
    'Opposition': Opposition,
    'Ground': Ground,
 #   'Economy': float(Economy),
    'Innings': int(Innings),
    'Overs': int(Overs)
}


index = [0]  # Index can be a list of integers
new_data = pd.DataFrame(new_data, index=index)


new_data['Opposition'] = labelencoder_1.transform(new_data['Opposition'])
new_data['Ground'] = labelencoder_2.transform(new_data['Ground'])
new_data['Innings'] = labelencoder_3.transform(new_data['Innings'])
new_data['Overs'] = labelencoder_4.transform(new_data['Overs'])
#new_data['Economy'] = labelencoder_5.transform(new_data['Economy'])


#labelencoder_1 = LabelEncoder()
#new_data['Opposition'] = labelencoder_1.fit_transform(new_data['Opposition'])

# Label encode the 'Ground' feature
#labelencoder_2 = LabelEncoder()
#new_data['Ground'] = labelencoder_2.fit_transform(new_data['Ground'])

#labelencoder_3 = LabelEncoder()
#new_data['Innings'] = labelencoder_3.fit_transform(new_data['Innings'])

#labelencoder_4 = LabelEncoder()
#new_data['Overs'] = labelencoder_4.fit_transform(new_data['Overs'])

#labelencoder_5 = LabelEncoder()
#new_data['Economy'] = labelencoder_5.fit_transform(new_data['Economy'])


print(new_data)

predicted_Wickets = model.predict(new_data)
print(f"Predicted Wickets for the future game against {Opposition} in {Ground}:", predicted_Wickets)
```
