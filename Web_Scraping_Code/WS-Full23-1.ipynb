```python
csv_file_path = '/content/drive/MyDrive/Capstone-Project/Player-Data/url-2023.csv'

urls = []
with open(csv_file_path, newline='') as csv_file:
  reader = csv.reader(csv_file)
  next(reader)  # Skip the header row
  for row in reader:
    url = row[0]
    urls.append(url)
```

```python
# Loop through the URLs and scrape the data
for url in urls:
    # Connect to the website and fetch the content
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all tables on the page
    tables = soup.find_all('table')

    # Create lists to hold the data
    dfs = []

    # Iterate over the tables and scrape the data
    for table in tables:
        table_data = []

        # Find all rows in the table and loop through them
        for row in table.findAll('tr'):
            row_data = []
            for cell in row.findAll(['td', 'th']):  # This will also fetch table headers
                row_data.append(cell.text)
            table_data.append(row_data)

        # Convert the list into a DataFrame
        df = pd.DataFrame(table_data)

        # Add the DataFrame to the list
        dfs.append(df)
```

```python
# Check URLs and filename
for df, url in zip(dfs, urls):
  print(url)
  filename = url.split('/')[-2].replace('.html', '.csv') + '.csv'
  print(filename)
  print(df)
```
